## Generate masked lung images
This notebook will generate masked lung images by applying the newly generated masks on its corresponding images.

```python
from src.preprocessing.masks_generator import generate_masked_images
from tensorflow.keras.models import load_model
import os

# Available categories in the dataset
#dataset = ['COVID','NORMAL','Viral Pneumonia','Lung_Opacity']
disease_classes = ['COVID','Normal','Viral Pneumonia','Lung_Opacity']

# images_data_dir = '../data/raw/dataset/augmented_images/'
# masks_data_dir = '../data/raw/dataset/masked_images/'
raw_img_dir = os.path.join("..","256x256_new_dataset_from_kaggle","raw")
masked_img_dir = os.path.join("..","256x256_new_dataset_from_kaggle","masked")

import os


# categories = ['Lung_Opacity','COVID', 'Normal', 'Viral Pneumonia', ]
categories = ['COVID', 'Normal', 'Viral Pneumonia', ]
for base_dir in [raw_img_dir, masked_img_dir]:
    # Create the base directory if it doesn't exist
    os.makedirs(base_dir, exist_ok=True)
    
    # Create subdirectories for each category
    for category in categories:
        folder_path = os.path.join(base_dir, category)
        os.makedirs(folder_path, exist_ok=True)
        print(f"Created or verified: {folder_path}")


# model_path = '../models/lung-segmentation-gan.h5'
model_path = os.path.join("..","models","lung-segmentation-gan.h5")
```

```python
# rename the files in the test dataset
import os

#raw_img_dir = os.path.join("..", "256x256_test_images_disease_prediction", "raw")
#masked_img_dir = os.path.join("..", "256x256_test_images_disease_prediction", "masked")
categories = ['COVID', 'Normal', 'Viral Pneumonia']

for category in categories:
    for img_type, base_dir in [('raw', raw_img_dir), ('masked', masked_img_dir)]:
        category_path = os.path.join(base_dir, category)
        if not os.path.exists(category_path):
            print(f"Skipping missing folder: {category_path}")
            continue

        files = sorted(os.listdir(category_path))
        for cnt, filename in enumerate(files, start=1):
            old_path = os.path.join(category_path, filename)
            if not os.path.isfile(old_path):
                continue
            _, ext = os.path.splitext(filename)
            new_filename = f"{category}_t_{cnt}{ext}"
            new_path = os.path.join(category_path, new_filename)

            os.rename(old_path, new_path)
            print(f"Renamed: {old_path} → {new_path}")
```

```python
# Load the pre-trained model

model = load_model(r"..\models\lung-segmentation-gan.h5", compile=False)
```

```python
for c in disease_classes:
    in_dir = os.path.join(raw_img_dir, c)
    out_dir = os.path.join(masked_img_dir, c)
    print(f"Generating masks for {in_dir} category...")
    generate_masked_images(in_dir, out_dir, model, target_size=(256, 256))
```
